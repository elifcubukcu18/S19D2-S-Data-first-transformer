### HuggingFace’i neden sevmelisiniz?

#### Önceden Eğitilmiş Modeller 📚

Transformers kütüphanesinin en güçlü özelliklerinden biri, sahip olduğu devasa **önceden eğitilmiş model** deposudur.  
BERT, GPT-2, T5, RoBERTa veya diğer transformer mimarilerini kullanmak istiyorsanız, model hub’da ihtiyacınıza uygun bir versiyon bulmanız çok olasıdır.

#### Çok Kolay 👍

Kütüphane, **kullanıcı dostu** olacak şekilde tasarlanmıştır.  
Bir modeli ve ona ait tokenizer’ı yüklemek yalnızca birkaç satır kod gerektirir. Aynı sadelik, **fine-tuning** sürecinde de geçerlidir; bu sayede güçlü modelleri çok farklı görevlere kolayca uyarlayabilirsiniz.

Kullanacağımız `pipelines` yapısı, model seçiminden sonuç almaya kadar olan süreci sadece birkaç satır kodla tamamlamanızı sağlar.

#### Tokenizer 🔄 ve Dataset 📊 Kütüphaneleri

Transformers kütüphanesine ek olarak HuggingFace:
- **Tokenizers** kütüphanesi ile hızlı ve kullanımı kolay tokenization yöntemleri,

sunar. Böylece ihtiyaç duyduğunuz tüm araçlar ve veriler tek bir ekosistemde toplanmış olur.

### Basic Sentiment : 😀 /  😕 / 😠 / 😟

Bunu göz önünde bulundurarak, bir model belirtmeden duygu analizi için bir pipeline oluşturun ve "Transformers are awesome!" cümlesiyle bu modeli test etmeye çalışın. Diğer cümleleri de denemekten çekinmeyin.

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

# test et
result = classifier("Transformers are awesome!")
print(result)
```

### Nüanslı Duygu 🤔

HuggingFace, bir model belirtmezsek varsayılan olarak `distilbert-base-uncased-finetuned-sst-2-english` modelini kullanır. Bu model birçok temel kullanım durumunda iyi sonuç verir, ancak oldukça sınırlı bir metin külliyatı üzerinde eğitildiği için: 

`Stanford Sentiment Treebank, dildeki duyguların kompozisyonel etkilerinin tam bir analizini mümkün kılan, tamamen etiketlenmiş ayrıştırma ağaçlarına sahip bir metin kümesidir. Bu metin kümesi, Pang ve Lee (2005) tarafından tanıtılan veri setine dayanmaktadır ve film eleştirilerinden çıkarılan 11.855 tek cümleden oluşmaktadır. Stanford ayrıştırıcı ile ayrıştırılmış ve bu ayrıştırma ağaçlarından toplam 215.154 benzersiz ifade içermektedir. Her ifade, 3 insan yargıç tarafından açıklanmıştır.`

Bu modelin, modern dil içeren cümlelerde muhtemelen kötü performans göstereceği oldukça açıktır: örneğin, "These beats are sick!" . Bu cümleleri şimdi boru hattınızda çalıştırmayı deneyin, oldukça olumlu duygular ifade etmelerine rağmen negatif puanlar alacaksınız.

```python
result2 = classifier('These beasts are good!')
print(result2)
```

HuggingFace modellerinin listesine gidin ve Twitter duygularını uzmanlaşmış bir model bulup bulamayacağınızı kontrol edin (`“twitter-roberta-base-sentiment-latest”` aramak iyi bir başlangıç olabilir) - umarım bu yeni terimler ile biraz daha güncel olur! Şimdi ikinci bir boru hattı oluşturun, bu sefer kullanmak istediğimiz modeli __belirtin__ (`model=` kullanın) ve ince ayarlı bir model kullandığımız için performansımızın anında nasıl iyileştiğini görün.

```python
classifier = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment"
)

result3 = classifier('These beasts are bad!')
```

```python
result3 = classifier('These beasts are bad!')
print(result3)
```

### Translation ✍️

Dil temamıza sadık kalalım ve bizim için bazı cümleleri çevirme görevlerini yerine getirebilecek bir model bulmaya çalışalım. Helsinki Üniversitesi'nin `opus-mt` projesi, HuggingFace'de inanılmaz derecede aktif olup, birçok farklı küresel dil için çeviri sürecini demokratikleştirmek üzere tasarlanmış modeller oluşturmakta ve sürdürmektedir. Şunu uygulamayı deneyin `"Helsinki-NLP/opus-mt-<source-language>-<destination-language>"` tiki dil arasında çeviri yapıp yapamayacağınızı görmek için (örneğin İngilizce'den İspanyolca'ya).

```python
pipe = pipeline('translation', model='Helsinki-NLP/opus-mt-ko-en')
result6 = pipe('이것은 솔직히 처음부터 끝까지 엄청난 실패였으며 젖은 비스킷의 모든 열정으로 연기한 일부 매우 보수가 좋은 배우들의 현저하게 영감을 받지 못한 연기로 끝났습니다.')
print(result6)
```

### Özetleme

Bir başka gerçekten yararlı NLP görevi, büyük miktarda bilgiyi çok az sayıda kelimeyle özetlemektir. BART, özetleme gibi görevlerde iyi performans gösteren bir modeldir; derslerde kısaca gördüğünüz iki modelin birleşiminden oluşur: BERT modeli ve otogresif stil GPT modeli. Bu model hakkında daha fazla bilgi için [bu bağlantıya](https://www.projectpro.io/article/transformers-bart-model-explained/553) göz atın. 

BART modelleri oldukça büyük olabileceğinden, HuggingFace'de mevcut en küçük damıtma modellerinden biri olan `distilbart-xsum-12-6` modelini bulmaya çalışın (damıtma hakkında daha sonra daha fazla konuşacağız!). Bu modeli bir `“özetleme”` boru hattına entegre edin, ardından bir metin alın (örneğin, [bir BBC makalesini](https://www.bbc.com/news/topics/cx2pk70323et) kopyalayıp yapıştırarak) ve boru hattınızla özetleyin!

Not: Bağlam pencerelerine dikkat etmelisiniz - burada, girişinizin model için çok uzun olması nedeniyle bir sorunla karşılaşabilirsiniz!

```python
text = """
A woman from Lincoln who lost more than 12 stone (76kg) in a year without using weight loss drugs is setting up her own slimming club to help others.
Jasmin Hyde, 30, previously weighed around 22 stone (139kg) but after starting a diet last April lost a total of 12 stone and 9lbs (about 80kg).
She said her success was down to "filling up on fresh foods rather than takeaways" and increasing the amount of exercise she did.
"There were definitely tough times, but I've had a hell of a lot of support around me, which keeps you motivated," she added.
Ms Hyde, who turned 30 in October, said she was motivated to do something about her weight after being diagnosed with polycystic ovary syndrome (PCOS) - which she said meant she may never be able to have her own children.
However, she said in order to get any support she would need to have a healthy body mass index (BMI).
She told BBC Radio Lincolnshire: "I thought I need to get cracked on - I'm turning 30 and I've got these issues."

"""
```

```python
pipe = pipeline("summarization", model="sshleifer/distilbart-xsum-12-6")
```

```python
result7 = pipe(text)
print(result7)
```
